                                       Supervised learning

Simple Linear Regression:


Independent variable(input): Hours studied because it's the factor we control or observe
Dependent Variable(output): Exam score because it depends on how many hours were studied

Importing the modules 

here we use the salary data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the Dataset
salary=pd.read_csv("salarydata.csv")
salary
```

```python
salary.info()
```

```python
#It will tell the statistics information about the dataset
salary.describe()
```

Segregate the dataset into independent variable(x) and dependent variable(y)

```python
x=salary.iloc[:,:-1].values
y=salary.iloc[:,1].values
```

                          Splitting the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split # here use module name as sklearn.mosel_selection & package is train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=2)
```

            Fitting the simple linear regression model to the training dataset

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

                   Prediction of Test and Training set Result

```python
x_pred=reg.predict(x_train)
y_pred=reg.predict(x_test)
```

```python
#TRAINING DATASET 
plt.scatter(x_train,y_train,color='green')
plt.plot(x_train,x_pred,color='red')
plt.title('Salary vs Experience (Training Dataset)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (in Rupees)')
plt.show()
```

```python
# TEST DATASET
plt.scatter(x_test,y_test,color='green')
plt.plot(x_train,x_pred,color='red')
plt.title('Salary vs Experience (Test Dataset)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (in Rupees)')
plt.show()
```

```python
#Prediction for all Test Data
y_predict=reg.predict(x_test)
y_predict.round()
```

```python
# Performance Analysis
from sklearn.metrics import mean_squared_error,r2_score
rscore=r2_score(y_test,y_predict)
r=rscore*100
print(round(r),'%')
```
